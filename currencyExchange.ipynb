```python
import requests as rq

cUrl = 'https://api.coinbase.com/v2/exchange-rates'
```

```python
cResp = rq.get(url=cUrl)
```

```python
cResp
```

```python
cResp.content
```

```python
cdecoded=cResp.json()
```

```python
cdecoded['data']['currency']
```

```python
cdecoded['data']['rates']['INR']
```

```python
inrparam = {
    'currency':'INR'
}

inrResp = rq.get(url=cUrl,params=inrparam)
```
